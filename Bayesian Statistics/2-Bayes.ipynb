 $$ \LARGE Bayesian \ Inference $$

$$ P(\theta|D) = \frac{P(D|\theta)P(\theta)}{P(D)} $$
$ Where:$
- $P(\theta)$ is the prior, the strength of our belief in $\theta$ without considering the evidence $D$. 

- $P(\theta|D)$ is the posterior, the refined strength of our belief in $\theta$ once the evidence $D$ has been taken into account. 

- $P(D|\theta)$ is the likelihood, the probability of observing the data $D$ given a model with parameter $\theta$. 

- $P(D)$ is the evidence, the probability of the data as determined by summing (or integrating) across all possible values of $\theta$, weighted by our beliefs in those particular values of $\theta$.



```python
# Bernoulli Distribution: coin flipping

import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

if __name__ == "__main__":
    number_of_trials = [0, 2, 10, 20, 50, 500]
    data = stats.bernoulli.rvs(0.5, size=number_of_trials[-1])
    x = np.linspace(0, 1, 100)
    for i, N in enumerate(number_of_trials):
        heads = int(data[:N].sum())
        ax = plt.subplot(len(number_of_trials) // 2, 2, i + 1)
        ax.set_title("%s trials, %s heads" % (N, heads))
        plt.xlabel("$P(H)$, Probability of Heads")
        plt.ylabel("Density")
        if i == 0:
            plt.ylim([0.0, 2.0])
        plt.setp(ax.get_yticklabels(), visible=False)
        y = stats.beta.pdf(x, 1 + heads, 1 + N - heads)
        plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
        plt.fill_between(x, 0, y, color="#aaaadd", alpha=0.5)
        
    plt.tight_layout()
    plt.show()
```
